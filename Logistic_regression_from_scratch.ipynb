Logistic regression from scratch

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Logistic Regression
class LogitRegression:
    def __init__(self, learning_rate, iterations):
        self.learning_rate = learning_rate
        self.iterations = iterations

    def fit(self, X, y):
        self.m, self.n = X.shape
        self.W = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.y = y

        for _ in range(self.iterations):
            self.update_weights()
        return self

    def update_weights(self):
        A = 1 / (1 + np.exp(-(self.X.dot(self.W) + self.b)))
        tmp = (A - self.y.T)
        tmp = np.reshape(tmp, self.m)
        dW = np.dot(self.X.T, tmp) / self.m
        db = np.sum(tmp) / self.m
        self.W = self.W - self.learning_rate * dW
        self.b = self.b - self.learning_rate * db
        return self

    def predict(self, X):
        Z = 1 / (1 + np.exp(-(X.dot(self.W) + self.b)))
        y_pred = np.where(Z > 0.5, 1, 0)
        return y_pred

# Data Cleaning
def clean_data(data):
    # No cleaning requirment no nulls , no one hot encodeing
    return data

# Split Dataset
def split_dataset(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Evaluation Metrics
def evaluate(y_true, y_pred):
    accuracy = np.mean(y_true == y_pred)
    precision = np.sum((y_true == 1) & (y_pred == 1)) / np.sum(y_pred == 1)
    recall = np.sum((y_true == 1) & (y_pred == 1)) / np.sum(y_true == 1)
    f1 = 2 * (precision * recall) / (precision + recall)
    return accuracy, precision, recall, f1

# Load dataset
def load_data(file_path):
    data = pd.read_csv(file_path)
    return data
```

```python
# Main function
def main():
    # Load data
    data = load_data("heart.csv")

    # Data Cleaning (if needed)
    data = clean_data(data)

    # Split dataset
    X = data.drop('target', axis=1).values
    y = data['target'].values
    X_train, X_test, y_train, y_test = split_dataset(X, y)

    # Train logistic regression model
    model = LogitRegression(learning_rate=0.1, iterations=2000000)
    model.fit(X_train, y_train)

    # Evaluate model
    y_pred = model.predict(X_test)
    accuracy, precision, recall, f1 = evaluate(y_test, y_pred)
    print("Accuracy on test set by logistic model:", accuracy * 100,"%")
    print("Precision on test set by logistic model:", precision)
    print("Recall on test set by logistic model:", recall)
    print("F1 Score on test set by logistic model:", f1)

if __name__ == "__main__":
    main()
```
